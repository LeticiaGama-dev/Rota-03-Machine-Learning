```python
from sklearn.model_selection import train_test_split

# ====== 1. Separar variáveis independentes (X) e variável dependente (y) ======
X = df.drop('Attrition', axis=1)  # todas as colunas, exceto a variável resposta
y = df['Attrition']               # variável resposta

# ====== 2. Dividir em treino e teste ======
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,          # 20% para teste
    random_state=42,         # garante reprodutibilidade
    stratify=y               # mantém a proporção da variável resposta
)

# ====== 3. Conferir resultados ======
print("Formato do conjunto de treino:", X_train.shape)
print("Formato do conjunto de teste:", X_test.shape)
print("\nProporção da variável resposta no conjunto de treino:")
print(y_train.value_counts(normalize=True))
print("\nProporção da variável resposta no conjunto de teste:")
print(y_test.value_counts(normalize=True))
```

```python
# ====== 1. Faixa etária ======
bins_age = [17, 25, 35, 45, 60]  # limites das categorias
labels_age = ['Jovem', 'Adulto', 'Sênior', 'Master']
X_train['AgeGroup'] = pd.cut(X_train['Age'], bins=bins_age, labels=labels_age)
X_test['AgeGroup'] = pd.cut(X_test['Age'], bins=bins_age, labels=labels_age)

# ====== 2. Razão de experiência na empresa ======
X_train['ExperienceRatio'] = X_train['YearsAtCompany'] / (X_train['TotalWorkingYears'] + 1)
X_test['ExperienceRatio'] = X_test['YearsAtCompany'] / (X_test['TotalWorkingYears'] + 1)

# ====== 3. Tempo desde última promoção (categorias) ======
bins_promo = [-1, 0, 3, 10, 20]  # limites das categorias
labels_promo = ['Recente', 'Média', 'Longa', 'Muito Longa']
X_train['PromotionGap'] = pd.cut(X_train['YearsSinceLastPromotion'], bins=bins_promo, labels=labels_promo)
X_test['PromotionGap'] = pd.cut(X_test['YearsSinceLastPromotion'], bins=bins_promo, labels=labels_promo)

# ====== 4. Número de empresas anteriores (categorias) ======
bins_companies = [-1, 1, 4, 10]  # limites das categorias
labels_companies = ['Pouca experiência', 'Experiência média', 'Muita experiência']
X_train['NumCompaniesGroup'] = pd.cut(X_train['NumCompaniesWorked'], bins=bins_companies, labels=labels_companies)
X_test['NumCompaniesGroup'] = pd.cut(X_test['NumCompaniesWorked'], bins=bins_companies, labels=labels_companies)

# ====== Conferir primeiras linhas ======
X_train[['Age', 'AgeGroup',
         'YearsAtCompany', 'TotalWorkingYears', 'ExperienceRatio',
         'YearsSinceLastPromotion', 'PromotionGap',
         'NumCompaniesWorked', 'NumCompaniesGroup']].head()
```

```python
# Selecionar colunas categóricas
categorical_cols = X_train.select_dtypes(include='object').columns
print("Colunas categóricas encontradas:\n", categorical_cols, "\n")

# Mostrar valores únicos de cada coluna categórica
for col in categorical_cols:
    print(f"{col} → valores únicos: {X_train[col].unique()}\n")
```

```python
# ====== 1️⃣ BusinessTravel (ordinal) ======
travel_map = {
    'Non-Travel': 0,
    'Travel_Rarely': 1,
    'Travel_Frequently': 2
}

X_train['BusinessTravelGroup'] = X_train['BusinessTravel'].map(travel_map)
X_test['BusinessTravelGroup'] = X_test['BusinessTravel'].map(travel_map)

# ====== 2️⃣ Department (Tech vs Non-Tech) ======
department_map = {
    'Research & Development': 'Tech',
    'Sales': 'Non-Tech',
    'Human Resources': 'Non-Tech'
}

X_train['DepartmentGroup'] = X_train['Department'].map(department_map)
X_test['DepartmentGroup'] = X_test['Department'].map(department_map)

# ====== 3️⃣ EducationField (áreas) ======
education_map = {
    'Life Sciences': 'Ciências',
    'Medical': 'Ciências',
    'Technical Degree': 'Técnico/Engenharia',
    'Marketing': 'Negócios',
    'Human Resources': 'Negócios',
    'Other': 'Outros'
}

X_train['EducationGroup'] = X_train['EducationField'].map(education_map)
X_test['EducationGroup'] = X_test['EducationField'].map(education_map)

# ====== 4️⃣ Gender (binário) ======
gender_map = {
    'Female': 0,
    'Male': 1
}

X_train['GenderGroup'] = X_train['Gender'].map(gender_map)
X_test['GenderGroup'] = X_test['Gender'].map(gender_map)

# ====== 5️⃣ JobRole (funções agrupadas) ======
jobrole_map = {
    'Research Director': 'Pesquisa',
    'Research Scientist': 'Pesquisa',
    'Laboratory Technician': 'Operacional',
    'Manager': 'Gestão',
    'Manufacturing Director': 'Gestão',
    'Healthcare Representative': 'Comercial',
    'Sales Representative': 'Comercial',
    'Sales Executive': 'Comercial',
    'Human Resources': 'RH'
}

X_train['JobRoleGroup'] = X_train['JobRole'].map(jobrole_map)
X_test['JobRoleGroup'] = X_test['JobRole'].map(jobrole_map)

# ====== 6️⃣ MaritalStatus (binário) ======
marital_map = {
    'Married': 1,
    'Single': 0,
    'Divorced': 0
}

X_train['MaritalStatusGroup'] = X_train['MaritalStatus'].map(marital_map)
X_test['MaritalStatusGroup'] = X_test['MaritalStatus'].map(marital_map)

# ====== Conferir primeiras linhas ======
X_train[['BusinessTravel', 'BusinessTravelGroup',
         'Department', 'DepartmentGroup',
         'EducationField', 'EducationGroup',
         'Gender', 'GenderGroup',
         'JobRole', 'JobRoleGroup',
         'MaritalStatus', 'MaritalStatusGroup']].head()
```

```python
# Bibliotecas necessárias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 1️⃣ Carregar dataset
df = pd.read_csv("/content/drive/MyDrive/ROTA 03/rh_data.csv")

# 2️⃣ Separar variável alvo e preditoras
y = df['Attrition'].map({'Yes':1, 'No':0})  # Transformar em 0 e 1
X = df.drop(columns=['Attrition', 'EmployeeID'])  # Tirar colunas irrelevantes

# 3️⃣ Identificar colunas numéricas e categóricas
numeric_cols = X.select_dtypes(include='number').columns
categorical_cols = X.select_dtypes(exclude='number').columns

# 4️⃣ Imputar valores ausentes
# Numéricas -> mediana
imputer_num = SimpleImputer(strategy='median')
X[numeric_cols] = imputer_num.fit_transform(X[numeric_cols])

# Categóricas -> moda
imputer_cat = SimpleImputer(strategy='most_frequent')
X[categorical_cols] = imputer_cat.fit_transform(X[categorical_cols])

# 5️⃣ Codificar variáveis categóricas
le = LabelEncoder()
for col in categorical_cols:
    X[col] = le.fit_transform(X[col])

# 6️⃣ Escalar variáveis numéricas
scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# 7️⃣ Divisão treino/teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 8️⃣ Definir modelos
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'RandomForest': RandomForestClassifier(n_estimators=200, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# 9️⃣ Treinar e avaliar modelos
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {
        'Acurácia': accuracy_score(y_test, y_pred),
        'Precisão': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1': f1_score(y_test, y_pred)
    }

# 10️⃣ Mostrar resultados
results_df = pd.DataFrame(results).T
print("=== Comparação de Modelos ===")
print(results_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# 1️⃣ Carregar dataset evoluído
df = pd.read_csv('/content/drive/MyDrive/ROTA 03/rh_data.csv')

# 2️⃣ Separar variáveis alvo e preditoras
target = 'Attrition'
X = df.drop(columns=[target, 'EmployeeID'])  # EmployeeID não contribui
y = df[target].map({'Yes':1, 'No':0})  # Converter para 0/1

# 3️⃣ Identificar colunas numéricas e categóricas
num_cols = X.select_dtypes(include='number').columns
cat_cols = X.select_dtypes(exclude='number').columns

# 4️⃣ Tratar valores ausentes
imputer_num = SimpleImputer(strategy='median')
X[num_cols] = imputer_num.fit_transform(X[num_cols])

imputer_cat = SimpleImputer(strategy='most_frequent')
X[cat_cols] = imputer_cat.fit_transform(X[cat_cols])

# Transformar categóricas em dummies
X = pd.get_dummies(X, drop_first=True)

# 5️⃣ Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 6️⃣ Treinar RandomForest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# 7️⃣ Obter importância das variáveis
importances = rf.feature_importances_
features = X.columns
feat_importance = pd.DataFrame({'Variável': features, 'Importância': importances})
feat_importance = feat_importance.sort_values(by='Importância', ascending=False)

# 8️⃣ Exibir
print("=== Importância das Variáveis ===")
print(feat_importance)

# 9️⃣ Plotar gráfico
plt.figure(figsize=(10,8))
plt.barh(feat_importance['Variável'], feat_importance['Importância'])
plt.gca().invert_yaxis()
plt.xlabel("Importância")
plt.title("Importância das Variáveis - Random Forest")
plt.show()
```
